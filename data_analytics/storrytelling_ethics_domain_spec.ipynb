### Load Data

```python
import pandas as pd

call_logging = pd.DataFrame({'Month':['January','February','March','April','May','June','July','August','September','October','November','December'],
                   'Ticket_Volume_Received':[160,184,241,149,180,161,132,202,160,139,149,177],
                   'Ticket_Volume_Processed':[160,184,237,148,181,150,123,156,126,104,124,140]})
names= ['January','February','March','April','May','June','July','August','September','October','November','December']
names2= ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

print(call_logging)
```

### "NORMAL" Plotting

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

sns.lineplot(x=call_logging.index,y=call_logging['Ticket_Volume_Received'],data=call_logging,marker='o',
             markersize=10,lw=4,label = 'Calls Logged',color='#4848FF',ax=ax)
sns.lineplot(x=call_logging.index,y=call_logging['Ticket_Volume_Processed'],data=call_logging,
             marker='s',markersize=10,lw=4,label = 'Calls Resolved',color='#FF6800',ax=ax)
plt.title('Monthly Volume of IT calls Logged and Resolved',fontsize=14,ha='right')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08),fancybox=True, shadow=False, ncol=5,fontsize=12)
plt.grid(axis='y')
plt.ylabel('')
plt.ylim(0,300)
ax.set(xticks=range(0, 12), xticklabels=list(names))

plt.show()
```

### Make it clean and simple

```python
%matplotlib inline
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

sns.lineplot(x=call_logging.index,y=call_logging['Ticket_Volume_Received'],data=call_logging,lw=4,color='#4848FF',ax=ax)
sns.lineplot(x=call_logging.index,y=call_logging['Ticket_Volume_Processed'],data=call_logging,lw=4,color='#FF6800',ax=ax)
plt.title('Monthly Volume of IT calls Logged and Resolved',fontsize=18,ha='center',color='#616161')
sns.despine()
ax.spines['left'].set_color('lightgray')
ax.spines['bottom'].set_color('lightgray')
ax.text(12.6, 180, 'Logged', ha='right', va='top', fontsize=22,color='#4848FF')
ax.text(12.8, 145, 'Resolved', ha='right', va='top', fontsize=22,color='#FF6800')

plt.xlabel('')
plt.ylim(0,300)
ax.set(xticks=range(0, 12), xticklabels=list(names2))
plt.ylabel('Volume',fontsize=16,color='#616161')

plt.show()
```
