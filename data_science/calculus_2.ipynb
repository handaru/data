### Get and visualize derivative of sine function, f(x) = sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
dy = np.cos(x)

print(f'derivative of sine function(-pi until pi): {dy}')

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.plot(x, dy, 'g', linewidth=2)
ax.legend(['y = sin(x)', "dy/dx = cos(x)"])
plt.show()
```

### Get integral sine function between pi/2 and pi

```python
from sympy import *

x = Symbol('x')
f = sin(x)
integral = integrate(f, (x, np.pi/2, np.pi))
print(integral)

# Create an array of x values
x = np.linspace(np.pi/2, np.pi, 100)
y = np.sin(x)

# Plot the function
plt.plot(x, y, label='sin(x)', color='red')

# Fill the area under the curve
plt.fill_between(x, y, where=(x >= np.pi/2) & (x <= np.pi), color='magenta', alpha=0.5, label='Area under curve')

# Add title and labels
plt.title('Integral of sin(x) from /2 to ')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.legend()

# Show the plot
plt.show()

```

### Get maximum of sine function between pi/2 and pi

```python
def f(x):
    return np.sin(x)

def grad(x):
    return np.cos(x)

x = np.pi
learning_rate = 0.1
iterations = 100

for i in range(iterations):
    x = x + learning_rate*grad(x)

print(x)
print(f(x))
```
